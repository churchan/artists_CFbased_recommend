### SVD

```python
## ## 先计算 artist_weights / 用户总weights 当做分值
```

```python
user_artists_sub_merged_weight_sum = user_artists_sub_merged[['userID','weight']].groupby('userID').sum().reset_index()
user_artists_sub_merged_weight_sum.head()
```

```python
user_artists_sub_merged_weight_sum.rename(columns={'weight':'total_weight'},inplace=True)
user_artists_sub_merged= pd.merge(user_artists_sub_merged,user_artists_sub_merged_weight_sum)
user_artists_sub_merged.head()
```

```python
user_artists_sub_merged['fractional_weights'] = user_artists_sub_merged['weight']/user_artists_sub_merged['total_weight']
user_artists_sub_merged[user_artists_sub_merged.userID ==2] 
```

```python
#### ## 建立用户、artists 和得分值的矩阵
```

```python
from scipy.sparse import coo_matrix

small_set = user_artists_sub_merged
user_codes = small_set.userID.drop_duplicates().reset_index()
artist_codes = small_set.userID.drop_duplicates().reset_index()
user_codes.rename(columns = {'index':'user_index'},inplace = True)
artist_codes.rename(columns = {'index':'artist_index'},inplace=True)
user_codes['so_index_value'] = list(user_codes.index)
artist_codes['us_index_value'] = list(artist_codes.index)
```

```python
small_set = pd.merge(small_set,user_codes,how = 'left')
small_set = pd.merge(small_set,artist_codes,how = 'left')
small_set.head()
```

```python
mat_candidate = small_set[['us_index_value','so_index_value','fractional_weights']]
data_array = mat_candidate.fractional_weights.values
row_array = mat_candidate.us_index_value.values
clo_array = mat_candidate.so_index_value.values

data_sparse = coo_matrix((data_array,(row_array,clo_array)),dtype=float)
```

```python
import math as mt
from scipy.sparse.linalg import *
from scipy.sparse.linalg import svds
from scipy.sparse import csc_matrix
```

```python
def compute_svd(urm,k):
    U,s,Vt = svds(urm,k)
    
    dim = (len(s),len(s))
    S = np.zeros(dim,dtype=np.float32)
    for i in range(0,len(s)):
        S[i,i] = mt.sqrt(s[i])
    
    U = csc_matrix(U,dtype=np.float32)
    S = csc_matrix(S,dtype=np.float32)
    Vt = csc_matrix(Vt,dtype=np.float32)
    
    return U,S,Vt

def compute_estimate_matrix(urm,U,S,Vt,Utest,K,best):
    rightItem = S*Vt
    max_recommendation = 50
    estimatedRatings = np.zeros(shape = (MAX_UID,MAX_PID),dtype=np.float16)
    recommendRatings = np.zeros(shape=(MAX_UID,max_recommendation),dtype=np.float16)
    for Usertest in Utest:
        prob = U[Usertest,:]*rightItem
        estimatedRatings[Usertest,:] = prob.todense()
        recommendRatings[Usertest,:] = (-estimatedRatings[Usertest,:]).argsort()[:max_recommendation]
    return recommendRatings
```

```python
k = 50
urm = data_sparse

MAX_UID = urm.shape[0]
MAX_PID = urm.shape[1]

U,S,Vt = compute_svd(urm,k)
```

### artists推荐

```python
## 随便挑几个用户id 进行推荐测试
Utest = [1,5,6,7]

Utest_recommended_items = compute_estimate_matrix(urm,U,S,Vt,Utest,k,True)
```

```python
for user in Utest:
    print("给ID为 {}号的用户进行音乐推荐". format(user))
    rank_value = 1
    for i in Utest_recommended_items[user,0:10]:
        artist_details = small_set[small_set.so_index_value == i].drop_duplicates('so_index_value')[['song','url']]
        print("第 {} 名推荐的音乐家是 {} ".format(rank_value, list(artist_details['song'])[0],list(artist_details['url'])[0]))
        rank_value+=1
```
